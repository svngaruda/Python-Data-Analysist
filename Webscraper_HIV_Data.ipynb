```python
import requests
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup as bs4
```

```python
hivaids_page = requests.get("https://id.wikipedia.org/wiki/Daftar_negara_menurut_prevalensi_HIV/AIDS_pada_orang_dewasa")
hivaids_soup = bs4(hivaids_page.content,'lxml')
```

```python
hivaids_table = hivaids_soup.find_all('table', {'class':'wikitable sortable'})
hivaids_table
```

```python
hivaids_df = pd.read_html(str(hivaids_table))
hivaids_df
```

```python
hivaids_df = hivaids_df[0]
hivaids_df = hivaids_df.rename(columns={"Country or region":"Country"})
hivaids_df
```
